```python
from tensorflow.keras.datasets import mnist # Importe le jeu de données MNIST (chiffres manuscrits) [cite: 41]
from tensorflow.keras.utils import to_categorical # [cite: 41]

# Charge les données MNIST en deux ensembles: entraînement (train) et test[cite: 41].
(X_train_data, Y_train_data), (X_test_data, Y_test_data) = mnist.load_data()

N = X_train_data.shape[0] # 60 000 données [cite: 41]
print(X_train_data[0].shape)
# print(X_train_data[0])

# vecteur image [cite: 41]
X_train = np.reshape(X_train_data, (N, 784))
X_test = np.reshape(X_test_data, (X_test_data.shape[0], 784))

X_train = X_train / 255 # normalisation [cite: 8]
X_test = X_test / 255 # normalisation [cite: 8]

print(X_train[0].shape)
print(X_test[0].shape)

Y_train = to_categorical(Y_train_data, num_classes=10) # [cite: 8]
Y_test = to_categorical(Y_test_data, num_classes=10) # [cite: 8]

print(Y_train[0])
print(Y_train_data[0])
```

```python
p = 8 # Nombre de neurones par couche [cite: 9]
modele = Sequential() # [cite: 9]

# Première couche : p neurones (entrée de dimension 784 = 28x28) [cite: 9]
modele.add(Dense(p, input_dim=784, activation='sigmoid'))

# Deuxième couche : p neurones [cite: 9]
modele.add(Dense(p, activation='sigmoid'))

# Couche de sortie : 10 neurones (un par chiffre) [cite: 9]
modele.add(Dense(10, activation='softmax'))

# Choix de la méthode de descente de gradient [cite: 9]
modele.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

print(modele.summary()) # [cite: 9]
```
