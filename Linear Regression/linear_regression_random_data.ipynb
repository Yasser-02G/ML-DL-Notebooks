```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title("Régression linéaire avec sklearn")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
print("Coefficient (a) :", model.coef_[0])    # pente
print("Intercept (b) :", model.intercept_)    # ordonnée à l'origine
print("Score R² :", model.score(x, y))    # qualité de prédiction
```
