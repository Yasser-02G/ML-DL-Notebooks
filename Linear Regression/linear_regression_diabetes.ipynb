```python
import numpy as np  # Bibliothèque pour le calcul scientifique
import matplotlib.pyplot as plt  # Bibliothèque de visualisation
from sklearn import datasets  # Fonctions pour charger des datasets classiques
from sklearn.linear_model import LinearRegression  # Implémentation de la régression linéaire
from sklearn.metrics import mean_squared_error, r2_score  # Métriques d'évaluation
```

```python
# Charger les données de diabète (sklearn.datasets)
diabetes = datasets.load_diabetes()

# On sélectionne uniquement la première variable (feature) pour simplifier
X = diabetes.data[:, np.newaxis, 0]  # Sélectionne la première colonne et ajoute une dimension
y = diabetes.target  # Variable cible (target)
```

```python
X_train = X[:-20]  # Toutes les données sauf les 20 dernières pour l'entraînement
X_test = X[-20:]   # Les 20 dernières données pour le test
y_train = y[:-20]  # Les targets correspondantes pour l'entraînement
y_test = y[-20:]   # Les targets correspondantes pour le test
```

```python
# Création et entraînement du modèle
model = LinearRegression()  # Initialisation du modèle
model.fit(X_train, y_train)  # Apprentissage sur les données d'entraînement
```

```python
y_pred = model.predict(X_test)  # Génère les prédictions pour les données de test
```

```python
# Création du graphique
plt.scatter(X_test, y_test, color='blue', label='Réel')  # Points bleus = données réelles
plt.plot(X_test, y_pred, color='red', label='Prédiction')  # Ligne rouge = prédictions du modèle
plt.title("Régression linéaire sur données réelles")  # Titre
plt.xlabel("Variable explicative")  # Label axe X
plt.ylabel("Valeur cible")  # Label axe Y
plt.legend()  # Affichage de la légende
plt.show()  # Affiche le graphique
```

```python
# Affichage des paramètres du modèle et des métriques de performance
print("Coefficient :", model.coef_)         # Affiche la pente de la droite de régression
print("Ordonnée à l'origine :", model.intercept_)  # Affiche le point d'intersection avec l'axe Y
print("MSE :", mean_squared_error(y_test, y_pred))  # Erreur Quadratique Moyenne
print("R2 score :", r2_score(y_test, y_pred))       # Coefficient de détermination
```

```python
plt.figure(figsize=(8, 5))
plt.plot(y_test, label="Valeurs réelles", marker='o')
plt.plot(y_pred, label="Valeurs prédites", marker='x')
plt.title("Comparaison des valeurs réelles et prédites")
plt.xlabel("Index")
plt.ylabel("Valeur cible")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
